shape 다루기

```python
# shape 출력하기
x = torch.randn(3,4)
x.shape
```

```python
# reshape - shape를 바꾸기. a*b = c*d 일때만 가능함
x = torch.randn(3,4)
x.reshape(2,6).shape
```

```python
# unsqueeze - 차원 추가
x = torch.randn(3)
print(x)
print(x.shape)

x_unsqueezed_0 = x.unsqueeze(0)

# unsqueeze하니까 껍데기가 하나 더 생김. shape 앞쪽에 3이 추가됨
# add a dimension / expand dims
# 0번 위치에 차원 추가. 그니까 3의 앞쪽 (0번인덱스)에 추가되는 거임
print(x_unsqueezed_0)
x_unsqueezed_0.shape
```

```python
x = torch.randn(3)
print(x)
print(x.shape)

x_unsqueezed_1 = x.unsqueeze(1)

# 이번엔 1번 위치에 차원 추가 -> 3의 뒤쪽에 추가됨
# 안쪽에 하나씩 껍데기가 생김
print(x_unsqueezed_1)
x_unsqueezed_1.shape
```

```python
# 반대로 squeeze의 경우 차원을 제거함.
x = torch.randn(1,3)

# 0번 차원을 제거해달라는 뜻
# 근데 여기서 0번 차원 크기가 1이면 제거되고, 1이 아니면 무시됨.
# x가 (2, 3)이었다면 무시되었을 것임!
x.squeeze(0).shape
```

```python
# permute - 텐서 순서 바꾸기
x = torch.randn(2,3,4)

# 0번은 그대로, 2번을 1번 자리로, 1번을 2번 자리로 바꾸기.
# ex. 이미지 텐서가 (배치, 높이, 너비, 채널)인데 (배치, 채널, 높이, 너비)로 바꾸고 싶다. 이럴 때 사용
x.permute(0, 2, 1).shape
```

연산 관련

# dim=0 0번차원 (행방향)합산. 열 끼리 더함

# dim=1 1번차원 (열방향)합산. 행끼리 더함

# "0열, 1행"

```python
x = torch.tensor(
    [[1., 2., 3.],
     [4., 5., 6.]]
)
# 열끼리 더한다.
# 1열 2열 3열 -> 5, 7, 9
x.sum(dim=0)
```

```python
# 행끼리 더한다.
# 1행
# 2행
# 3행
# 1+2+3, 4+5+6
x.sum(dim=1)
```

```python
# 열끼리 평균.
# 1열, 2열, 3열
# 1+4 / 2 ...
x.mean(dim=0).values

# 각 열에서 제일 큰값.
# 1열에서 max, 2열에서 max, ...
x.max(dim=0).values
```

```python
# @ 는 행렬곱. weighted sum에 사용됨!
# (3*4) * (4*5) = (3*5)
a = torch.randn(3,4)
b = torch.randn(4,5)
(a @ b).shape
```

```python
# 그냥 * 하면 원소별로 곱함
# 같은 위치에 있는 애들끼리 곱해줌. shape는 그대로!!
a = torch.tensor([1., 2., 3.])
b = torch.tensor([4., 5., 6.])
a * b
```
