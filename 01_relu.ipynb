## ReLU
Rectified Linear Unit (= 정규화된 선형 유닛)

f(x) = max(0, x)
입력값이 0보다 크면 그대로 통과, 0 이하라면 0으로 만듦

왜 렐루를 쓸까?
- x>0 인지 확인만 하는 수준이라 연산이 빠름. 대규모모델 학습 시 장점
- vanishing gradient 해결 - 양수일때 기울기가 항상 1이기 때문
- sparsity 생성 - 0 이하 값들을 모두 0으로 만들기 때문에 뉴런 중 일부만 활성화됨 -> 계산 효율성 up

Dying ReLU = 렐루 함수는 입력값이 0 이하일때 기울기가 0임. 그래서 어떤 뉴런이 계속 음수 값만 받으면 그 뉴런은 영원히 0만 출력하고 함수가 멈춰버림.
-> 보완하기 위해 0 이하일때 0 대신에 아주 작은 값을 주는 Leaky ReLU 같은 함수도 나옴


````python
# ✏️ YOUR IMPLEMENTATION HERE

def relu(x: torch.Tensor) -> torch.Tensor:
    # if 를 사용하면 텐서의 모든 원소를 하나하나 체크해야 함.
    # 그래서 if문 대신 mask를 만들어서, 행렬곱으로 한번에 처리!
    mask = (x > 0).float()
    return x * mask
````

````python
# 🧪 Test your implementation (feel free to add more debug prints)
x = torch.tensor([-2., -1., 0., 1., 2.])
print("Input: ", x)
print("Output:", relu(x))
print("Shape: ", relu(x).shape)
````
