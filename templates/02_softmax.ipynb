# 🟢 Easy: Implement Softmax

Implement the **Softmax** function from scratch.

$$\text{softmax}(x_i) = \frac{e^{x_i}}{\sum_j e^{x_j}}$$

### Signature
```python
def my_softmax(x: torch.Tensor, dim: int = -1) -> torch.Tensor:
    ...
```

### Rules
- Do **NOT** use `torch.softmax`, `F.softmax`, or `torch.nn.Softmax`
- Must be **numerically stable** (hint: subtract `max` before `exp`)

### Example
```
Input:  tensor([1., 2., 3.])
Output: tensor([0.0900, 0.2447, 0.6652])  # sums to 1.0
```

````python
# ✏️ YOUR IMPLEMENTATION HERE

"""
dim, keepdim에 대해
왜 max, sum에만 쓰냐? 이 함수들은 여러 값을 합쳐서 하나로 줄이는 연산이기 때문
shape가 바뀌게 되니까
어느방향으로 줄일지(dim), 줄인 후 차원을 유지할지(keepdim)을 알려줘야 함!
exp의 경우 각 원소를 독립적으로 변환하기 때문에 이 두개가 불필요함
"""

def my_softmax(x: torch.Tensor, dim: int = -1) -> torch.Tensor:
    # softmax의 성질: 모든 값에 같은 상수 더하거나 빼도 결과 같음!
    # 제일 큰 값을 0으로 맞춰줌.
    # ex. x = [1000, 999, 998] -> [0, -1, -2]
    # 이렇게 하면 overflow 방지

    # x내의 모든 값들에 대해 dim 방향으로 최댓값을 구함
    # keepdim=True - broadcasting
    x_max = x.max(dim=dim, keepdim=True).values
    # x안의 모든 값들에 대해 e^x 해줌. 그리고 새로운 텐서에 저장
    e_x = torch.exp(x - x_max)
    return e_x / e_x.sum(dim=dim, keepdim=True)
````

````python
# 🧪 Debug
x = torch.tensor([1.0, 2.0, 3.0])
print("Output:", my_softmax(x, dim=-1))
print("Sum:   ", my_softmax(x, dim=-1).sum())  # should be ~1.0
print("Ref:   ", torch.softmax(x, dim=-1))
````
